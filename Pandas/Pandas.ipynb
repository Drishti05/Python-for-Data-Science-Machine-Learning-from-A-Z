## Creating a database

```python
import pandas as pd
import numpy as np
```

```python
data={'Date':['2016-03-01','2016-03-02','2016-03-01','2016-03-03','2016-03-02','2016-03-03'],'Type':['a','b','c','a','a','c'],'Value':[11.432,13.031,20.784,99.906,1.303,20.784]}
```

```python
df=pd.DataFrame(data,columns=['Date','Type','Value'])
```

```python
df
```

```python
s= pd.Series([3,-5,7,4],index=['a','b','c','d'])
```

```python
s
```

# Reshaping Data

## Pivot

```python
df1=df.pivot(index='Date',columns='Type',values='Value')
```

```python
df1
```

## Pivot Table

```python
df4=pd.pivot_table(df,values='Value',index='Date',columns='Type')
```

```python
df4
```

## Stack / Unstack

```python
stacked = df.stack()
```

```python
stacked
```

```python
stacked.unstack()
```

## Melt

```python
pd.melt(df,id_vars=['Date'],value_vars=['Type','Value'],value_name='Observations')
```
